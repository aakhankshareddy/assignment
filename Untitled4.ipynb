```python
from transformers import pipeline
translation=pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
prompt = "Hi !How are you?"
print(translation(prompt))
```

```python
from transformers import pipeline
translation=pipeline("translation_en_to_es", model="google/flan-t5-base")
prompt = """ Give me only the answer:
English: Good Morning
Spanish: Buen dia
English: Thank You
Spanish: """
print(translation(prompt))
```

```python
from transformers import pipeline
cal=pipeline("text2text-generation", model="google/flan-t5-large")
prompt="If a bag has 25 chocolates and 10 are distributed, how many are left? Only answer."
print(cal(prompt))
```
